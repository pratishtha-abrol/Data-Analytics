#### Concept Heirarchy Table
A concept hierarchy is a sequence of mappings from a set of low-level concepts to higher-level, more general concepts. Many of these are implicit in the database schema. (*Country*, *City* and *Locality_name* can be calculated through *Latitute* and *Longitute* information, boolean variables *is_Apartment* and *is_Studio* are derived from the attribute *Property_type*) 

Concept hierarchies may also be defined by discretizing or grouping values or a given dimension, resulting in a set-grouping hierarchy.  

A concept hierarchy that is a total or partial order among the attributes in a database schema is called schema hierarchy.

In a multi-dimensional OLAP model, data are organized into multiple dimensions and each dimension contains multiple levels of abstraction defined by concept hierarchies. This organisation allows users the flexibility to view data from different perspectives. A number of OLAP data cube operations exist to materialize these different views, allowing interactive querying and analysis of the data at hand. 

Concept Heierarchy is already included in the given database.
